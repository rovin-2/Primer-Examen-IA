```python
import pandas as pd
import numpy as np
```

```python
vdf = pd.read_csv('fruitvegprices-2017_2022.csv')
```

```python
vdf['date'] = pd.to_datetime(vdf['date'])
```

## Ejercicios

### Ejercicio uno: realice un analisis de valores nulos

En caso de existir valores nulos elimine esos registros

```python
analic_nulos = pd.DataFrame()
```

```python
for anio in vdf.loc[ : , "date Year" ].drop_duplicates():
    vdf_year = vdf.loc[ vdf['date Year'] == anio , : ]    
    for columna in vdf_year.columns:        
        porcentage =  np.sum(pd.isna( vdf_year.loc[ : , columna ]  )) / vdf_year.shape[0]
        nue_df = pd.DataFrame({
            'year': [ anio ]
            , 'column' : [ columna ]
            , 'percentage de nulos' : [ porcentage ]
        })
        analic_nulos = pd.concat([analic_nulos, nue_df])
```

```python
analic_nulos
```

### Ejercicio dos: Por categoria, cuantos distintos items hay y cuantos distintos variety hay?

El resultado debe de tener el siguiente formato de presentacion:

```python
pd.DataFrame({
    "category" : ['cut_flowers','fruit']
    , "item_count" : [9,10]
    , "variety_Count" : [10,21]
})
```

Como nota en un mismo dataframe mostramos todo lo solicitado, usted debe de mostrar todas las categorias

```python
nuevo_vdf = vdf.loc[ :, ["category","item","variety"]].drop_duplicates()
```

```python
nuevo_vdf = nuevo_vdf.groupby(["category"], as_index=False).agg({"item":["count"],"variety":["count"]})
```

```python
nuevo_vdf.columns = nuevo_vdf.columns.droplevel(0)
```

```python
nuevo_vdf.columns = ["category","item_count","variety_Count"]
```

### Ejercicio cuatro: Segun el ejercicio tres, que tanto aumento o disminuyo el precio de cada item-variety 

Ejemplo de como se debe de mostrar su respuesta

```python
pd.DataFrame({
    "item" : ['Manzana','Manzana','Manzana']
    , "variety" : ['Gala','Gala','Gala']
    , "year" : [2020,2021,2022]
    , "last_price" :  [2.7,2.5,3.2]
    , "delta" : [0, (2.5-2.7), (3.2-2.5)]
})
```

Segun nuestro ejemplo dummy para 2021 estuvo 0.2 centavos mas barato y en 2022 estuvo 0.7 centavos caro

```python
ultimo_valor_vdf = ultimo_valor_vdf.sort_values(["item","variety"])
```

```python
ultimo_valor_vdf["delta"] = ultimo_valor_vdf.groupby(["item","variety","date Year"], as_index=False)["last price"].shift(1)
```

```python
ultimo_valor_vdf = ultimo_valor_vdf.fillna(0)
```

```python
ultimo_valor_vdf["disminuyo_aumento"] = (ultimo_valor_vdf["last price"] - ultimo_valor_vdf["delta"])
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
porcertanje_vdf = vdf.loc[ :, ["category","item","variety","price"] ]
```

```python
porcertanje_vdf = porcertanje_vdf.groupby(["category","item","variety"], as_index=False).agg({"price":["max","min"]})
```

```python
porcertanje_vdf.columns = porcertanje_vdf.columns.droplevel(0)
```

```python
porcertanje_vdf.columns = ["category","item","variety","precio_maximo","precio_minimo"]
```

```python
porcertanje_vdf["percentage-change-between-first-and-last-price"] = (porcertanje_vdf["precio_maximo"] - porcertanje_vdf["precio_minimo"])/100
```
